> 방학 계획 | 한글 폰트 깨짐 영어로 표현

```python
import matplotlib.pyplot as plt
# 키값 저장
activities = [
    'Sleeping (24:00~9:00)',
    'Breakfast (9:00~10:00)',
    'Studying (10:00~12:00)',
    'Cleaning (12:00~13:00)',
    'Solving Algorithms (13:00~15:00)',
    'Japanese & Math Study (15:00~17:00)',
    'Coding Development (17:00~22:00)',
    'YouTube (22:00~24:00)'
]
hours = [9, 1, 2, 1, 2, 2, 5, 2]  # 각 활동의 시간 (총합 24)
start_angle = 90  # 90도로 설정하면 12시 방향 시작
plt.figure(figsize=(10, 10))
wedges, texts, autotexts = plt.pie(
    hours,
    labels=activities,
    autopct='%1.1f%%',
    startangle=start_angle,
    counterclock=False
)
for text in texts:
    text.set_fontsize(10)
for autotext in autotexts:
    autotext.set_fontsize(10)
plt.title('Vacation Plan - 12:00 Start', fontsize=16)
plt.axis('equal')
plt.show()
```
